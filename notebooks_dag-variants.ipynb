# Random Devils Eating Groundhog function
Start by importing the random package:

```python
import random
```

Create the lists to pull randomly from:

```python
devils=[
    "Satan",
    "Lucifer",
    "Beelzebub",
    "Mephistopheles",
    "Asmodeus",
    "Belial",
    "Mammon",
    "Ahriman",
    "Iblis",
    "Shaitan",
    "Old Nick",
    "Prince of Darkness",
    "The Horned One"
]

groundhogs=[
    "the woodchuck",
    "the whistlepig",
    "the groundhog",
    "the marmot",
    "the land-beaver",
    "the rockchuck",
    "the burrower",
    "the tunnel digger",
    "the furry prophet"
]

verbs=[
    "ate",
    "devoured",
    "consumed",
    "chowed down on",
    "gobbled up"
]
```

Define a function called "rdag()" which creates the string:

```python
def rdag():
    return f"{random.choice(devils)} {random.choice(verbs)} {random.choice(groundhogs)}"
```
